```python
import numpy as np
import pandas as pd

import pulp
# print(pulp.__doc__)

from pulp import LpProblem, LpVariable, LpMaximize,LpMinimize, value, lpSum, lpDot, LpBinary, LpInteger, LpStatus
import dual

from ortoolpy import addvar, addvars, addbinvar, addbinvars,addvals, model_min, model_max

from itertools import product

import random
```

```python
n_job = 10
sample_lb_list = [1, 2, 3, 4, 4, 6, 5, 2, 2, 1]
sample_ub_list = [1, 2, 3, 4, 4, 6, 5, 2, 2, 1]

n_person = 30

sample_lb_dict = {}
sample_ub_dict = {}
for i in range(n_job):
    sample_lb_dict['job_' + str(i)] = sample_lb_list[i]
    sample_ub_dict['job_' + str(i)] = sample_ub_list[i]

wish_list_dict = {}

for i in range(n_person):
    person_wish_list = []
    
    for job in range(n_job):
        if random.random() >= 0.50:
            person_wish_list.append(random.randint(0, 100))
        else:
            person_wish_list.append(0)
    
    person_wish_list = list(map(lambda x:round((x / sum(person_wish_list))*100, 1), person_wish_list))
    wish_list_dict['person_' + str(i)] = person_wish_list

index_job_list = ['job_' + str(i) for i in range(n_job)]
    
wish_list_dict
index_job_list
import_df = pd.DataFrame(wish_list_dict, index=index_job_list)
import_df = import_df.T
display(import_df)
display(sample_lb_dict)
display(sample_ub_dict)
```

```python
df = import_df.copy()
df = df.replace({0: -100000,})
df
```

```python
prob = model_max()

lb_for_job_dict = sample_lb_dict
ub_for_job_dict = sample_ub_dict
x = {}
p = {}

for i, j in product(df.index, df.columns):
    x[i,j] = LpVariable('x('+str(i)+','+str(j)+')', lowBound=0)
    p[i,j] = df.loc[i,j]
    
prob += lpSum(p[i,j]*x[i,j] for i,j in product(df.index, df.columns))

for i in df.index:
    prob += lpSum(x[i,j] for j in df.columns) == 1

for j in df.columns:
    prob += lpSum(x[i,j] for i in df.index) >= lb_for_job_dict[j]
    prob += lpSum(x[i,j] for i in df.index) <= ub_for_job_dict[j]

prob.solve()

print(LpStatus[prob.status])
print(value(prob.objective))
```

```python
dfr = df.copy()
satisfaction_level_list = []
satisfaction_rank_list = []
assigned_department_list = []
for i in df.index:
    satisfaction_level_list.append(sum([value(p[i,j]*x[i,j]) for j in df.columns]))
    satisfaction_rank_list.append(sum([df.rank(ascending=False, axis=1).loc[i,j]*value(x[i,j]) for j in df.columns]))
    for j in df.columns:
        if value(x[i,j]) == 1:
            assigned_department_list.append(j)
        
dfr['希望度'] = satisfaction_level_list
dfr['希望順位'] = satisfaction_rank_list
dfr['配属先'] = assigned_department_list
```
